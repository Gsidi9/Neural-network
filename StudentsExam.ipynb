```python
import numpy as np 
```

# Define Independent varibale and dependent variable

```python
#Indepedent variable/ values that the model takes as an input, suing them to predict the dependent variable
input_set = np.array([[5,7,1],
                [8,9,5],
                [10,4,4],
                [0,6,2],
                [1,9,8],
                [2,3,7]])


#Dependent variable/ The value the model is aim to predict
y = np.array([[0,
               1,
               1,
               0,
               1,
              0]])
    
#converts the answers to vector
y= y.reshape(6,1)
```

Our input set contains six records of data.Similary,I also created a set (y) that contains the corresponding labels in the input set. the set(y) are the vakues we want our model to predict

# Define Hyperparameters

```python
np.random.seed(0)
# setting up the weight to a random number
#since we have three features in the input set , we have a vector of three weights
weights = np.random.rand(3,1)
# setting up the bias to a random number
bias = np.random.rand(1)
#The amount that the weight is updated during training is called learning rate,
lr= 0.05
```

# Define the Activation function

```python
#our activation function is sigmoid function
def sigmoid(x):
    return 1/(1+np.exp(-x))
# define a function that calculates the derivate fo the sigmoid function
def sigmoid_deviration(x):
    return sigmoid(x)*(1-sigmoid(x))
```

# Train the model

```python
for epoch in range(10000): # number of time I want to train the algorithm on our datasets
#storing the values of input_sets to the inputs variaables, so that the input_setsremain as it is in each iteration 
    inputs = input_set
    #find the dot product of the inputs and the weight and add the bias to it
    XW = np.dot(inputs, weights) +bias
    # We pass the dot product summaion through the sigmoid action function/ The variable z contains the predicted ouputs
    z = sigmoid(XW)
    
    #computes error rate for back-propagation
    error = z - y
    print(error.sum())
    #dcost = error
    dpred = sigmoid_deviration(z)
    #performing weight adjustments
    z_del = error * dpred
    inputs = inputs.T
    weights = weights - lr*np.dot(inputs, z_del)
    
    #multiply the learning rate variable lr with the derivative to increase the speed of learning.Alsothe bias term.
    for num in z_del:
        bias = bias - lr*num
    
```

Once the loop starts, the total errors starts decreasing and by the end of the trainingthe error is left with a very small value

# Make predictions with new data

```python
train = np.array([3,9,10])
result = sigmoid(np.dot(train, weights)+ bias)
print("output:",result)
```

The output is closer to 1 , so it is classified as pass.
