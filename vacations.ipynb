```python
import pandas as pd
import numpy as np
import glob
```

```python
filenames = glob.glob('*.csv')
df=[]
for file in filenames:
    print('Reading'+ file)
    df.append(pd.read_csv(file))
```

```python
df = pd.concat(df)
```
